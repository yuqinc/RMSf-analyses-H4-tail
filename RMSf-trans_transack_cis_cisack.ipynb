##  Impact of trans and cis lesion on the H4 tail RMSf 

```python
fig, ax = plt.subplots(figsize=(16,10))
sns.pointplot(x="resi", y="trans",  data=df, color="blue", markers=["o"],linestyles=["-"], figsize=(12,6),label='trans')
#sns.pointplot(x="resi", y="transack",label="trans_ack", data=df, color="orange",markers=["o"],linestyles=["-"], figsize=(12,6))
sns.pointplot(x="resi", y="cis",  data=df, color="purple", markers=["o"],linestyles=["-"],label='cis', figsize=(12,6))
sns.pointplot(x="resi", y="unmod", data=df, color="grey",markers=["o"],linestyles=["--"], label="unmod",figsize=(12,6))
plt.legend()

sns.set_context("notebook", font_scale=2, rc={"lines.linewidth": 2})
ax.set_xlabel('H4 tail residue number')
ax.set_ylabel('RMSf (Angstrom)')
plt.savefig("trans_cis_unmod.png")
```

## Impact of the acetylation on the tran_BP containing NCP H4 tail

```python
fig, ax = plt.subplots(figsize=(16,10))
sns.pointplot(x="resi", y="cis", label='cis', data=df, color="pink", markers="o",linestyles="-", figsize=(12,6))
sns.pointplot(x="resi", y="cis_ack",label="cis_ack", data=df, color="green",markers="o",linestyles="-", figsize=(12,6))
sns.pointplot(x="resi", y="unmod", data=df, color="grey",markers=["o"],linestyles=["--"], label="unmod",figsize=(12,6))

plt.legend()
sns.set_context("notebook", font_scale=2, rc={"lines.linewidth": 2})
ax.set_xlabel('H4 tail residue number')
ax.set_ylabel('RMSf (Angstrom)')
plt.savefig("cis_cisack.png")
```

```python
fig, ax = plt.subplots(figsize=(16,10))
sns.pointplot(x="resi", y="trans", label="trans", data=df, color="blue", markers="o",linestyles="-", figsize=(12,6))
sns.pointplot(x="resi", y="transack",label="trans_ack", data=df, color="orange",markers="o",linestyles="-", figsize=(12,6))
sns.pointplot(x="resi", y="unmod", data=df, color="grey",markers=["o"],linestyles=["--"], label="unmod",figsize=(12,6))

plt.legend()
sns.set_context("notebook", font_scale=2, rc={"lines.linewidth": 2})
ax.set_xlabel('H4 tail residue number')
ax.set_ylabel('RMSf (Angstrom)')
plt.savefig("trans_transack.png")
```
